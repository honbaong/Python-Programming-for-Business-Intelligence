## Task 1: A list of dictionaries (1p)

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

#### a) Filter out invalid transactions where the price is zero or the quantity is negative.

```python
valid_tx = [tx for tx in transactions if tx["price"] > 0 and tx["quantity"] > 0]
```

#### b) For each remaining transaction, compute a new field total = price * quantity.

```python
for tx in valid_tx:
    tx["total"] = tx["price"] * tx["quantity"]
```

#### c) Group the valid transactions by region and compute the total revenue per region. Store the result in a dictionary.

```python
from collections import defaultdict
revenue_by_region = defaultdict(int)
for tx in valid_tx:
    revenue_by_region[tx["region"]] += tx["total"]
```

#### d) Print the region with the highest total revenue and its value using an f-string.

```python
max_region = max(revenue_by_region, key=revenue_by_region.get)
print(f"Region with highest revenue: {max_region}, Total: {revenue_by_region[max_region]}")
```

## Task 2: Creating a function (1p)

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales_list):
    total = sum(sales_list)
    average = round(total / len(sales_list), 2)
    best_day = sales_list.index(max(sales_list)) + 1
    worst_day = sales_list.index(min(sales_list)) + 1
    above_avg_count = sum(1 for s in sales_list if s > average)
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_avg_count
    }

print(summarise_sales(sales))
```

## Task 3: Loops (1p)

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

#### a) Create a new list called passing that contains only the names of students with a grade of 50 or above. Print the list.

```python
passing = [s["name"] for s in students if s["grade"] >= 50]
print("Passing students:", passing)
```

#### b) Loop through the students and print a status message for each. If the student's grade is 80 or above, print "[name]: Excellent"; if between 50 and 79, print "[name]: Pass"; otherwise print "[name]: Fail".

```python
for s in students:
    if s["grade"] >= 80:
        print(f"{s['name']}: Excellent")
    elif s["grade"] >= 50:
        print(f"{s['name']}: Pass")
    else:
        print(f"{s['name']}: Fail")
```

#### c) Count how many students have an attendance rate below 0.80. Print the count.

```python
low_attendance_count = sum(1 for s in students if s["attendance"] < 0.8)
print("Students with attendance < 0.8:", low_attendance_count)
```

#### d) Find the student with the highest grade using a loop (do not use max()). Print the student's name and grade.

```python
highest_grade_student = students[0]
for s in students[1:]:
    if s["grade"] > highest_grade_student["grade"]:
        highest_grade_student = s
print("Highest grade (loop):", highest_grade_student["name"], highest_grade_student["grade"])
```

#### e) Now find the student with the highest grade using max(). Print the student's name and grade.

```python
highest_grade_student_max = max(students, key=lambda s: s["grade"])
print("Highest grade (max):", highest_grade_student_max["name"], highest_grade_student_max["grade"])
```

## Task 4: Class and methods (1p)

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}

    def total_value(self):
        return sum(v["shares"] * v["price_per_share"] for v in self.holdings.values())

    def most_valuable(self):
        return max(self.holdings, key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"])

    def __str__(self):
        return f"Portfolio of {self.owner}: {len(self.holdings)} assets, Total value: {self.total_value():.2f}"

# Example usage
pf = Portfolio("Alice")
pf.add_asset("AAPL", 10, 150)
pf.add_asset("TSLA", 5, 700)
pf.add_asset("GOOG", 8, 1200)

print(pf)
print("Most valuable asset:", pf.most_valuable())
print("Total portfolio value:", pf.total_value())
```

## Task 8: More Pandas (1p)

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

#### a) Create a DataFrame with columns: date, visitors, conversion_rate, and avg_order_value. Set date as the index.

```python
df_traffic = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df_traffic.set_index("date", inplace=True)
```

#### b) Compute two new columns: conversions (= visitors × conversion_rate, rounded to the nearest integer) and daily_revenue (= conversions × avg_order_value).

```python
df_traffic["conversions"] = (df_traffic["visitors"] * df_traffic["conversion_rate"]).round().astype(int)
df_traffic["daily_revenue"] = df_traffic["conversions"] * df_traffic["avg_order_value"]

df_traffic.head()
```
